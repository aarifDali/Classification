```python
import pandas as pd
```

```python
df = pd.read_csv('HRDataset_v14.csv')
```

```python
name = []
F_Name =[]
L_Name = []
for i in df['Employee_Name']:
    name = str.split(i, ',')
    F_Name.append(name[0])
    L_Name.append(name[1])
       
```

```python
df['F_Name'] = F_Name
df['L_Name'] = L_Name
```
